```python
from exa_py import Exa

exa = Exa(api_key=os.environ['EXA_API_KEY'])


def search_interests(query, category=None):
    response = exa.search_and_contents(
        query,
        use_autoprompt=True,
        num_results=5,
        category=category,
        text={"max_characters": 1000},
        highlights=True
    )
    
    for result in response.results:
        print(f"Title: {result.title}")
        print(f"URL: {result.url}")
        print(f"Highlight: {result.highlights[0] if result.highlights else 'N/A'}")
        print("---")

# Example searches
search_interests("Latest machine learning tutorials")
search_interests("Popular Python coding projects", category="github")
search_interests("Interesting TED talks on technology", category="videos")
```
